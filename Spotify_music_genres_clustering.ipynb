```python
import pandas as pd
import numpy as np
from sklearn import cluster
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import plotly.graph_objects as go
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv('Spotify-2000.csv')
data.head()
```

```python
#drop the index column
data.drop("Index", axis=1)
```

```python
#correlation between features
data.corr()
```

```python
#clustering analysis
#select relevant features from data
data2 = data[["Beats Per Minute (BPM)", "Loudness (dB)", 
              "Liveness", "Valence", "Acousticness", 
              "Speechiness"]]
```

```python
#scale data
for i in data.columns:
    MinMaxScaler(i)

#fit data to the kmeans model
kmeans = KMeans(n_clusters=10)
clusters = kmeans.fit_predict(data2)
```

```python
#add the predicted clusters to the dataset
data["Music Segments"] = clusters
MinMaxScaler(data["Music Segments"])
data["Music Segments"] = data["Music Segments"].map({1: "Cluster 1", 2: 
    "Cluster 2", 3: "Cluster 3", 4: "Cluster 4", 5: "Cluster 5", 
    6: "Cluster 6", 7: "Cluster 7", 8: "Cluster 8", 
    9: "Cluster 9", 10: "Cluster 10"})
```

```python
data.head()
```

```python
#visualize based on some features [Liveness, Popularity, Loudness ]
PLOT = go.Figure()
for i in list(data["Music Segments"].unique()):
    

    PLOT.add_trace(go.Scatter3d(x = data[data["Music Segments"]== i]['Liveness'],
                                y = data[data["Music Segments"] == i]['Popularity'],
                                z = data[data["Music Segments"] == i]['Loudness (dB)'],                        
                                mode = 'markers',marker_size = 6, marker_line_width = 1,
                                name = str(i)))
PLOT.update_traces(hovertemplate='Liveness: %{x} <br>Popularity: %{y} <br>Loudness: %{z}')

    
PLOT.update_layout(width = 1800, height = 900, autosize = True, showlegend = True,
                   scene = dict(xaxis=dict(title = 'Liveness', titlefont_color = 'black'),
                                yaxis=dict(title = 'Popularity', titlefont_color = 'black'),
                                zaxis=dict(title = 'Loudness (dB)', titlefont_color = 'black')),
                   font = dict(family = "Gilroy", color  = 'black', size = 12))
```
