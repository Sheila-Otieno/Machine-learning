```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Johnny Depp","shooting","f1"])
data = trends.interest_by_region()
data = data.sort_values(by="Johnny Depp",ascending=False)
data = data.head(25)
print(data)
```

```python
data.reset_index().plot(x="geoName", y=["Johnny Depp","shooting","f1"], 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data2 = trends.interest_over_time()
data2.head()
```

```python
data2.reset_index().plot(x="date", y=["Johnny Depp","shooting","f1"], 
                        figsize=(20,15), kind="bar")
plt.show()
```

```python
kw_list=["fenty","spotify","johnny depp","monkeypox","f1"]
data3 = trends.get_historical_interest(kw_list, year_start=2022, month_start=1, day_start=1, hour_start=0, year_end=2022, month_end=2, day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=0)
data3.head()
```

```python
data3.reset_index().plot(x="date", y=["fenty","spotify","johnny depp","monkeypox","f1"], 
                        figsize=(20,15), kind="bar")
plt.show()
```

```python
data4 = TrendReq(hl='en-US', tz=155)
data4.build_payload(kw_list=["fenty","spotify","johnny depp","monkeypox","f1"])
data4 = data4.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data4["fenty"].plot()
data4["spotify"].plot()
data4["johnny depp"].plot()
data4["monkeypox"].plot()
data4["f1"].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for trending topics in Nairobi', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.legend()
plt.show()
```
