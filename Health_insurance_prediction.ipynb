```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv("Health_insurance.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
#Distribution of people who smoke and those that do not
data = data
figure = px.histogram(data, x = "sex", color = "smoker", title= "Number of Smokers")
figure.show()
```

```python
#change categorical data into 1 and 0
data["sex"] = data["sex"].map({"female": 0, "male": 1})
data["smoker"] = data["smoker"].map({"no": 0, "yes": 1})
data.head()
```

```python
#distribution of people living in each region
pie = data["region"].value_counts()
regions = pie.index
population = pie.values
fig = px.pie(data, values=population, names=regions)
fig.show()
```

```python
#data correlation among features
data.corr()
```

```python
x = np.array(data[["age", "sex", "bmi", "smoker"]])
y = np.array(data["charges"])


xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Random forest
forest = RandomForestRegressor()
forest.fit(xtrain, ytrain)
```

```python
#predicted values of the model
ypred = forest.predict(xtest)
data = pd.DataFrame(data={"Predicted Premium Amount": ypred})
data.head(10)
```
