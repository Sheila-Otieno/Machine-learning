```python
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

# **Load Summarization pipeline**

```python
summarizer = pipeline('summarization')
```

# **Get a blog post**
Blog post from Bikozulu

```python
URL ="https://bikozulu.co.ke/arm-and-a-dress/"
```

```python
##use request to get information from the URL
r = requests.get(URL)
```

```python
r.text
```

```python
#Beautiful soup is used to extract text from blog with H1 tag and p tag

soup = BeautifulSoup(r.text, 'html.parser')
results = soup.find_all(['h1','p'])
```

```python
results
```

```python
#concatenate all the text into one block of text

text = [result.text for result in results]
article = ' '.join(text)
```

```python
article
```

### Chunk text


Split text into individual sentences

## Remove fullstops and replace them with end of text tag **bold text**

```python
article = article.replace('.', '.<eos>')
article = article.replace('?', '?<eos>')
article = article.replace('!', '!<eos>')
sentences = article.split('<eos>')
```

```python
sentences[1]
```

```python
#limit the max chunk to 500
#loop through each sentence in order to split it into chunks
max_chunk = 500 
current_chunk = 0 
chunks = []
for sentence in sentences:
    if len(chunks) == current_chunk + 1: 
        if len(chunks[current_chunk]) + len(sentence.split(' ')) <= max_chunk:
            chunks[current_chunk].extend(sentence.split(' '))
        else:
            current_chunk += 1
            chunks.append(sentence.split(' '))
    else:
        print(current_chunk)
        chunks.append(sentence.split(' '))
```

```python
chunks[0]
```

```python
#join the individual chunks into one chunk
for chunk_id in range(len(chunks)):
    chunks[chunk_id] = ' '.join(chunks[chunk_id])
```

```python
chunks[4]
```

```python
len(chunks[4].split(' '))
```

```python
len(chunks)
```

# **Summarize text**

```python
res = summarizer(chunks, max_length=120, min_length=30, do_sample=False)
```

```python
len(res)
```

```python
' '.join([summ['summary_text'] for summ in res])
```

```python
#join all the 13 individual summaries into one summary
text = ' '.join([summ['summary_text'] for summ in res])
```

```python
text
```

## **Output to a text file**

```python
with open('blogsummary.txt', 'w') as f:
    f.write(text)
```
